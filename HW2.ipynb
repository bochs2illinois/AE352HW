```python
#Question 1
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
import math

m = 4 #kg
k = 10 #n/m
d = 0.5 #m

#a
wnf = np.sqrt(k/m)
print("natural frequency: ", wnf, "s-1")

#b
y, cc = symbols("y, cc")
eq1 = Eq(0, (k/m) - y**2)
eq2 = Eq(y, cc / (2*m))
soln = solve((eq1,eq2),(y,cc))
c = soln[1][1]/10

w0 = math.sqrt((k/m) - (c/(2*m))**2)
print("frequency with damping is ", w0, "s-1")

#c
wf = w0
print("Drive it a wf = w0, ", wf, "s-1")

#d
a = (1 / math.sqrt( ((m**2)*((k/m)-(wf**2))**2) + ((c**2)*(wf**2)) )) * d

print("The amplitude at resonance is: ", a)
```

```python
#Question 3
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp


#given
mc = 12 #amu
mo = 16 #amu
k = 1860 #N/m
amu = 1.66 * (10**(-27)) #kg

mc = mc*amu
mo = mo*amu

#b
M = np.array([[-k/mc, k/mc], [k/mo, -k/mo]])
eigvals, eigvecs = np.linalg.eig(M)
print(eigvals)
print(eigvecs)
eigfreq = np.sqrt(-eigvals[0])
print("The eigenfrequency of the system is: ", eigfreq, "hz") 
#c

#d
c = 299792458 #m/s
v = eigfreq/(2*np.pi)
wavelength =  c/v
print("The wavelength to excite the molecules is: ", wavelength, "meters")
```

```python
#Question 4
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy.integrate import solve_ivp


#given
m1 = 1000 #kg
m2 = 2000 #kg
m3 = 3000 #kg
k1 = 3000 #N/m
k2 = 2000 #N/m
k3 = 1000 #N/m
x10 = 1 #m
x20 = 0 #m
x30 = 0 #m

M = np.array([[-(k1+k2)/m1, k2/m1, 0],
              [k2/m2, -(k2+k3)/m2, k3/m2],
              [0, k3/m3, -k3/m3]])

eigvals, S = np.linalg.eig(M)


print(eigvals)
print(S)

eigfreq = np.sqrt(-eigvals)
print(eigfreq)


t = np.linspace(0, 20, 10000)  
q1 = S[0][0]*np.cos(eigfreq[0]*t)
q2 = S[1][0]*np.cos(eigfreq[1]*t)
q3 = S[2][0]*np.cos(eigfreq[2]*t)

plt1 = q1
plt.xlim([0, 20])
plt.ylim([-2, 2])
plt.grid(True, linestyle =':')
plt.plot(t, plt1, 'r' )
plt.xlabel('Time')
plt.ylabel('Displacement in Meters')
plt.title('Eigenmode q1')
plt.show()

plt2 = q2
plt.xlim([0, 20])
plt.ylim([-0.5, 0.5])
plt.grid(True, linestyle =':')
plt.plot(t, plt2, 'g' )
plt.xlabel('Time')
plt.ylabel('Displacement in Meters')
plt.title('Eigenmode q2')
plt.show()

plt3 = q3
plt.xlim([0, 20])
plt.ylim([-0.1, 0.1])
plt.grid(True, linestyle =':')
plt.plot(t, plt3, 'b' )
plt.xlabel('Time')
plt.ylabel('Displacement in Meters')
plt.title('Eigenmode q3')
plt.show()
```
