```python
#Question 1
import numpy as np
from sympy import *
import matplotlib.pyplot as plt

#given
h0 = 10000 #m
vz = -100 #m/s
r = 1000 #m
T = 6*np.pi #s

#setup
t = symbols("t")
w = (1/T) * np.pi *2


#Part A

#position in cartesian coordinates
x = Matrix([r*cos(w*t),r*sin(w*t),h0+vz*t])

#velocity in cartesian coordinates
v = diff(x,t)

#acceleration in cartesian coordinates
A = diff(v,t)

#jerk in cartesian coordinates
j = diff(A,t)

print("Position in meters: ")
pprint(x)
print("Velocity in meters per second: ")
pprint(v)
print("Acceleration in meters per second per second: ")
pprint(A)
print("Jerk in meters per second per second per second: ")
pprint(j)

#Part B
pltx = 7
plty = 4
#Position
x1 = np.linspace(0, 120, 1000)
 
fig = plt.figure(figsize = (pltx, plty))
 
# Plot X Pos
y = r*np.cos(w*x1)
plt.plot(x1, y, 'r', label ='X-Position')

# Plot Y Pos
y1 = r*np.sin(w*x1)
plt.plot(x1, y1, 'g', label ='Y-Position')

# Plot Z Pos
y2 = h0+vz*x1
plt.plot(x1, y2, 'b', label ='Z-Position')
 
 
plt.legend()
plt.grid(True, linestyle =':')
plt.xlim([0, 100])
plt.ylim([-1200, 12000])
 
plt.title('X, Y, and Z Position')
plt.xlabel('Time in Seconds')
plt.ylabel('Displacement in Meters')
plt.show()

#Velocity
x2 = np.linspace(0, 100, 1000)
 
fig1 = plt.figure(figsize = (pltx, plty))
 
# Plot X Vel
v = -333.333333333333*np.sin(0.333333333333333*x2)
plt.plot(x2, v, 'r', label ='X-Velocity')

# Plot Y Vel
v1 = 333.333333333333*np.cos(0.333333333333333*x2)
plt.plot(x2, v1, 'g', label ='Y-Velocity')

# Plot Z Vel
v2 = -100+(.000000000001*x2)
plt.plot(x2, v2, 'b', label ='Z-Velocity')
 
 
plt.legend()
plt.grid(True, linestyle =':')
plt.xlim([0, 100])
plt.ylim([-400, 500])
 
plt.title('X, Y, and Z Velocity')
plt.xlabel('Time in Seconds')
plt.ylabel('Velocity in Meters per Second')
plt.show()

#Acceleration
x3 = np.linspace(0, 100, 1000)
 
fig1 = plt.figure(figsize = (pltx, plty))
 
# Plot X Acc
a = -111.111111111111*np.cos(0.333333333333333*x3)
plt.plot(x3, a, 'r', label ='X-Acceleration')

# Plot Y Acc
a1 = -111.111111111111*np.sin(0.333333333333333*x3)
plt.plot(x3, a1, 'g', label ='Y-Acceleration')

# Plot Z ACc
a2 = 0*x2
plt.plot(x3, a2, 'b', label ='Z-Acceleration')
 
 
plt.legend()
plt.grid(True, linestyle =':')
plt.xlim([0, 100])
plt.ylim([-150, 200])
 
plt.title('X, Y, and Z Acceleration')
plt.xlabel('Time in Seconds')
plt.ylabel('Acceleration in Meters per Second per Second')
plt.show()

#Part C
amag = A.norm()
Amag = amag.subs(t,0)
print("The maximuim acceleration is ", Amag, "meters per second per second. This is over 11 G and will make the passengers and pilot black out!")
```

```python
#Question 2
import numpy as np
from sympy import *

#given
T = 115000 #lbf
m = 80 #kg 
CS = 7.5 #ft2
h = 1 #m

#setup
T = 4.44822*T #conversion to N
CS = 0.092903*CS #conversion to m2

#Part A
a = T/m
print("Acceleration: ", a, " m/s/s")

#Part B
t, x = symbols("t, x")
A = Matrix([a,-9.81])
V = integrate(A,t)
D = integrate(V,t) + Matrix([0,h])

eq1 = Eq(0, D[1])
eq2 = Eq(x, D[0])

soln = solve((eq1,eq2),(t, x))
print("The person is flung back ", soln[1][1], " meters.")
```

```python
#Question 3
import numpy as np
from sympy import *

#given
h = 3000 #m
m = 75 #kg
A = 0.18 #m2
Cd = 0.7
p = 1.2 #kg/m3

#setup
g = -9.81

#Part B
Vt, D, Fg = symbols("Vt, D, Fg")

eq1 = Eq(D, 0.5*p*(Vt**2)*Cd*A)
eq2 = Eq(Fg, m*g)
eq3 = Eq(Fg, -D)

soln = solve((eq1,eq2,eq3),(Vt,D,Fg))
print("Terminal velocity is ", soln[1][0], " m/s down")

#Part C
t = 1/100 #s
DP = soln[1][0]*m - 0*m
F = DP/t
a = F/m
print("Acceleration at impact: ", a, " m/s/s")
```

```python
#Question 4
import numpy as np
from sympy import *


#given
hp = 1000 #m
m = 75 #kg
Ap = 20 #m2
Cdp = 1.75
p = 1.2 #kg/m3\
vT = -100

#setup
g = -9.81

#Part A
Vtp, Dp, Fgp = symbols("Vtp, Dp, Fgp")

eq1 = Eq(Dp, 0.5*p*(Vtp**2)*Cdp*Ap)
eq2 = Eq(Fgp, m*g)
eq3 = Eq(Fgp, -Dp)

soln = solve((eq1,eq2,eq3),(Vtp,Dp,Fgp))
newtv = round(soln[1][0],12)
print("Terminal velocity is ", newtv, " m/s down")

#Part B
v = symbols("v")
dvdt = ((1/2)*p*v**2*Cdp*Ap)/m + g

h = 0
t = 0
dt = .001

while vT < -newtv:
    a = dvdt.subs(v,vT)
    vT = vT + a*dt
    h = h + vT*dt + (1/2)*a*dt**2
    t = t + dt

print("The time to deccelerate is ", t, " seconds.")
print("The distance you fall in this time in ", -h, "meters")

#Part C
vtp = soln[1][0]
ti = 1/6 #s
DPp = soln[1][0]*m - 0*m
Fi = DPp/ti
a = Fi/m
print("Acceleration at impact: ", a, " m/s/s")
```

```python
# Question 5
import numpy as np
from sympy import *
from scipy.integrate import odeint
import matplotlib.pyplot as plt


#given
h = 3000 #m
m = 75 #kg
A = 0.18 #m2
Cd = 0.7
p = 1.2 #kg/m3
hd = 1000 #m
cw = 100 #km/hr East
hw = 500 #m
gs = 200 #km/hr North
hp = 1000 #m
Ap = 20 #m2
Cdp = 1.75
g = 9.81
As = 5 #m2

#setup
cw = cw*1000/60/60 #Conversion m/s
gs = gs*1000/60/60 #Conversion m/s


#Part A
# Z direction, Altitude
#3-1 km Z
def model(z1, t1):
    u1, v1 = z1
    dz1dt1 = [v1, (((1/2)*p*(v1**2)*Cd*A)/m) - g]
    return dz1dt1
z10 = [3000, 0]
t1 = np.linspace(0, 27.25, 10000)  
Z1 = odeint(model, z10, t1)


#1-0km Z
def model(z2, t2):
    u2, v2 = z2
    dz2dt2 = [v2, (((1/2)*p*(v2**2)*Cdp*Ap)/m) - g]
    return dz2dt2
z20 = [1000, -98.6516236923795]
t2 = np.linspace(27.25, 195, 10000)  
Z2 = odeint(model, z20, t2)



#X Direction, +East
#0.5-0 km X
def model(x1, t3):
    u3, v3 = x1
    dx1dt3 = [v3, (((1/2)*p*((cw-v3)**2)*Cdp*As)/m)]
    return dx1dt3
x10 = [0, 0]
t3 = np.linspace(110.5, 195, 100000)  
X1 = odeint(model, x10, t3)

t4 = np.linspace(0, 110.5, 10)  
#3-0.5 km X
x2 = 0*t4
plt.plot(t4, x2, 'r')

#Plot X
pltx1 = X1[:, 0]
plt.xlim([0, 200])
plt.ylim([0, 2700])
plt.grid(True, linestyle =':')
plt.plot(t3, pltx1, 'r', label='X(t)')
plt.xlabel('Time')
plt.ylabel('Displacement in Meters')
plt.title('X Displacement, +East')
plt.legend()
plt.show()

#Y Direction, +North
#3-1 km Y
def model(y1, t5):
    u4, v4 = y1
    dxydt5 = [v4, -(((1/2)*p*((v4)**2)*Cdp*A)/m)]
    return dxydt5
y10 = [0, gs]
t5 = np.linspace(0, 27.25, 100000)  
Y1 = odeint(model, y10, t5)

#1-0km Y
def model(y2, t6):
    u5, v5 = y2
    dxydt6 = [v5, -(((1/2)*p*((v5)**2)*Cdp*As)/m)]
    return dxydt6
y20 = [Y1[100000-1][0], Y1[100000-1][1]]
t6 = np.linspace(27.25, 195, 100000)  
Y2 = odeint(model, y20, t6)


#Plot Y
plty1 = Y1[:, 0]
plty2 = Y2[:, 0]
plt.xlim([0, 200])
plt.ylim([0, 800])
plt.grid(True, linestyle =':')
plt.plot(t5, plty1, 'g', label='Y(t)')
plt.plot(t6, plty2, 'g')
plt.xlabel('Time')
plt.ylabel('Displacement in Meters')
plt.title('Y Displacement, +North')
plt.legend()
plt.show()

#Plotting Z
pltz1 = Z1[:, 0]
pltz2 = Z2[:, 0]
plt.xlim([0, 200])
plt.ylim([0, 3100])
plt.grid(True, linestyle =':')
plt.plot(t1, pltz1, 'b', label='Z(t)')
plt.plot(t2, pltz2, 'b')
plt.xlabel('Time')
plt.ylabel('Height in Meters')
plt.title('Z Displacement, Height')
plt.legend()
plt.show()

#Part B
print("If the I consider the planes ground speed before jump I would have drifted ", pltx1[100000-1], "meters off the expected landing spot.")
print("The total dispacement from the spot on the ground above the jump point would be ", np.sqrt(pltx1[100000-1]**2 + plty2[100000-1]**2), " meters.")
```
