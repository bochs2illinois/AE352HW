```python
#Question 5
%reset
from sympy import *
import numpy as np
import matplotlib.pyplot as plt

#paramaters
t = symbols("t")

MI = Matrix([[600,0,0], 
             [0, 500,0], 
             [0,0, 500]])

T_max = 60*10
dt = 0.01

#Rotation Setup
#Quaternians
q0     = Function('q0')(t)
q1     = Function('q1')(t)
q2     = Function('q2')(t)
q3     = Function('q3')(t)

q = Matrix([q0,q1,q2,q3])

phi   = atan2(2*(q0*q1 + q2*q3), q0**2 - q1**2 - q2**2 + q3**2)
theta = asin(2*(q0*q2 - q3*q1))
psi   = atan2(2*(q0*q3 + q2*q1), q0**2 + q1**2 - q2**2 - q3**2)

wx    = Function('wx')(t)
wy    = Function('wy')(t)
wz    = Function('wz')(t)

wxd   = diff(wx,t)
wyd   = diff(wy,t)
wzd   = diff(wz,t)

w     = Matrix([wx,wy,wz])

wd    = Matrix([wxd,wyd,wzd])

R = Matrix([[1-2*q2**2-2*q3**2      ,         2*q1*q2-2*q0*q3,         2*q1*q3+2*q0*q2],
            [        2*q1*q2+2*q0*q3, 1-2*q1**2-2*q3**2      ,         2*q2*q3-2*q0*q1],
            [        2*q1*q3-2*q0*q2,         2*q2*q3+2*q0*q1, 1-2*q1**2-2*q2**2      ]])

wd = MI.inv() @  Matrix([0,0,0]) - MI.inv() @ (w.cross(MI @ w))

ic_q0     = 1  
ic_q1     = 0  
ic_q2     = 0
ic_q3     = 0 

ic_wx     = .1   
ic_wy     = .2    
ic_wz     = .1    

#Simply Propagator

T         = np.array([0])

q0_t      = np.array([ic_q0])
q1_t      = np.array([ic_q1])
q2_t      = np.array([ic_q2])
q3_t      = np.array([ic_q3])

wx_t      = np.array([ic_wx])
wy_t      = np.array([ic_wy])
wz_t      = np.array([ic_wz])

q0d_t      = np.array([-0.5*(-ic_wx*ic_q1 - ic_wy*ic_q2 - ic_wz*ic_q3)])
q1d_t      = np.array([-0.5*( ic_wx*ic_q0 + ic_wy*ic_q3 - ic_wz*ic_q2)])
q2d_t      = np.array([-0.5*(-ic_wx*ic_q3 + ic_wy*ic_q0 + ic_wz*ic_q1)])
q3d_t      = np.array([-0.5*( ic_wx*ic_q2 - ic_wy*ic_q1 + ic_wz*ic_q0)])

wxd_t     = np.array(  [wd[0].subs(q0,ic_q0).subs(q1,ic_q1).subs(q2,ic_q2).subs(q3,ic_q3).subs(wx,ic_wx).subs(wy,ic_wy).subs(wz,ic_wz)])
wyd_t     = np.array(  [wd[1].subs(q0,ic_q0).subs(q1,ic_q1).subs(q2,ic_q2).subs(q3,ic_q3).subs(wx,ic_wx).subs(wy,ic_wy).subs(wz,ic_wz)])
wzd_t     = np.array(  [wd[2].subs(q0,ic_q0).subs(q1,ic_q1).subs(q2,ic_q2).subs(q3,ic_q3).subs(wx,ic_wx).subs(wy,ic_wy).subs(wz,ic_wz)])

p = 0

while (T[p] < T_max):
    wx_t     = np.append(wx_t,      wx_t[p]      + wxd_t[p]    * dt)
    wy_t     = np.append(wy_t,      wy_t[p]      + wyd_t[p]    * dt)
    wz_t     = np.append(wz_t,      wz_t[p]      + wzd_t[p]    * dt)
    
    q0d_t    = np.append(q0d_t,   -0.5*(-wx_t[p]*q1_t[p] - wy_t[p]*q2_t[p] - wz_t[p]*q3_t[p])) 
    q1d_t    = np.append(q1d_t,   -0.5*( wx_t[p]*q0_t[p] + wy_t[p]*q3_t[p] - wz_t[p]*q2_t[p])) 
    q2d_t    = np.append(q2d_t,   -0.5*(-wx_t[p]*q3_t[p] + wy_t[p]*q0_t[p] + wz_t[p]*q1_t[p]))
    q3d_t    = np.append(q3d_t,   -0.5*( wx_t[p]*q2_t[p] - wy_t[p]*q1_t[p] + wz_t[p]*q0_t[p])) 
    
    q0_t     = np.append(q0_t,     q0_t[p]      + q0d_t[p]     * dt) 
    q1_t     = np.append(q1_t,     q1_t[p]      + q1d_t[p]     * dt) 
    q2_t     = np.append(q2_t,     q2_t[p]      + q2d_t[p]     * dt) 
    q3_t     = np.append(q3_t,     q3_t[p]      + q3d_t[p]     * dt) 
        
    wxd_t    = np.append(wxd_t,    wd[0].subs(q0,q0_t[p+1]).subs(q1,q1_t[p+1]).subs(q2,q2_t[p+1]).subs(q3,q3_t[p+1]).subs(wx,wx_t[p+1]).subs(wy,wy_t[p+1]).subs(wz,wz_t[p+1]))
    wyd_t    = np.append(wyd_t,    wd[1].subs(q0,q0_t[p+1]).subs(q1,q1_t[p+1]).subs(q2,q2_t[p+1]).subs(q3,q3_t[p+1]).subs(wx,wx_t[p+1]).subs(wy,wy_t[p+1]).subs(wz,wz_t[p+1]))
    wzd_t    = np.append(wzd_t,    wd[2].subs(q0,q0_t[p+1]).subs(q1,q1_t[p+1]).subs(q2,q2_t[p+1]).subs(q3,q3_t[p+1]).subs(wx,wx_t[p+1]).subs(wy,wy_t[p+1]).subs(wz,wz_t[p+1]))
    
    T = np.append(T,T[p] + dt)
    
    if p % 1000 == 0:
        print(p)
    p = p + 1 
    

```

```python
#quaternians to euler
phi_t   = np.array([])
theta_t = np.array([])
psi_t   = np.array([])
T_      = np.array([])

i = 0

while i < len(q0_t) :
    phi_t = np.append(phi_t, re(phi.subs(q0,q0_t[i]).subs(q1,q1_t[i]).subs(q2,q2_t[i]).subs(q3,q3_t[i])))
    theta_t = np.append(theta_t, re(theta.subs(q0,q0_t[i]).subs(q1,q1_t[i]).subs(q2,q2_t[i]).subs(q3,q3_t[i])))
    psi_t = np.append(psi_t, re(psi.subs(q0,q0_t[i]).subs(q1,q1_t[i]).subs(q2,q2_t[i]).subs(q3,q3_t[i])))
    
    T_ = np.append(T_, T[i])
    
    i = i + 10
    if i % 1000 == 0:
        print(i)

for i in range(len(phi_t)):
    if phi_t[i] == nan:
        phi_t[i] = 0
    if theta_t[i] == nan:
        theta_t[i] = 0
    if psi_t[i] == nan:
        psi_t[i] = 0
```
